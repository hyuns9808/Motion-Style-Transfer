#  Motion Retargeting - Bandico

From paper: https://deepmotionediting.github.io/retargeting

Cross-Structural Motion Retargeting:
Retarget motion to asymmetric characters or character with missing/extra bones.
The motions are performed by skeletons from another homeomorphic
domain

Intra-Structural Retargeting:
Retargeting of skeletons with the same structure, but different proportions.
The motions are performed by skeletons within the same domain.

We will run cross-structural retargeting on the Bandico dataset referencing the Mixamo dataset, as our goal is to create a dataset that uses the SAME primal skeleton with the Maximo dataset.
